## Load Waymo package

```python
from waymo_open_dataset.utils import frame_utils
from waymo_open_dataset import dataset_pb2 as open_dataset
import tensorflow as tf
```

## Load HiddenFootprints package

```python
from hiddenfootprints import core as hiddenfootprints_func
```

## Read frames (image, label, camera params)

```python
# load one tfrecord file from the Waymo dataset
waymo_dir = '/datasets/waymo_raw/training_0000/' # <-- change to your waymo directory
tfrecord_filename = waymo_dir+'segment-10023947602400723454_1120_000_1140_000_with_camera_labels.tfrecord'
```

```python
segment = tf.data.TFRecordDataset(tfrecord_filename, compression_type='')

# camera
cameras = [
    'FRONT',
    'FRONT_LEFT',
    'FRONT_RIGHT',
    'SIDE_LEFT',
    'SIDE_RIGHT'
]

cameras_img_size = {
    'FRONT': (1280, 1920, 3),
    'FRONT_LEFT': (1280, 1920, 3),
    'FRONT_RIGHT': (1280, 1920, 3),
    'SIDE_LEFT': (886, 1920, 3),
    'SIDE_RIGHT': (886, 1920, 3)
}

# select camera
selected_camera = 'FRONT'

# load image, labels, camera parameters and store them in 'frames'
frames = {}
for (frame_idx,data) in enumerate(segment):
    frame_record = open_dataset.Frame()
    frame_record.ParseFromString(bytearray(data.numpy()))

    frames[frame_idx] = hiddenfootprints_func.read_single_frame(
                        frame_record, open_dataset, selected_camera
                        )
```

## Hidden Footprints projection from all frames to reference frame

```python
# reference frame
reference_frame_idx = 120

# get propagated labels
reference_frame_propagated_labels = hiddenfootprints_func.propagate(reference_frame_idx, frames)
```

```python
# Visualization

# pick one class
# 1 for cars
# 2 for pedestrians
# 3 for signs
# 4 for cyclists
picked_labels_boolean = reference_frame_propagated_labels[:,0]==2
picked_labels = reference_frame_propagated_labels[ picked_labels_boolean ]
print('{} propagated labels. '.format(picked_labels_boolean.sum()))

import matplotlib.pyplot as plt
import matplotlib.patches as patches

def show_camera_image_with_points(image, points):
  """Show a camera image and the given points."""    
  assert len(points.shape)==2 and points.shape[1]==2, 'points should be Nx2'

  ax = plt.subplot(1,1,1)
  for point in points:
      # Draw the object bounding box.
      ax.add_patch(patches.Rectangle(
        xy=(point[0],
            point[1]),
        width=8,
        height=8,
        linewidth=3,
        edgecolor='green',
        facecolor='none'))

  # Show the camera image.
  plt.imshow(image)
  plt.grid(False)
  plt.axis('off')    
    
plt.figure(figsize=(25,20));show_camera_image_with_points(frames[reference_frame_idx]['im'], picked_labels[:,1:])
```

## Load the pretrained model to predict walkable regions

```python
from hiddenfootprints.model.networks import FootprintsPredictor

model_file = './WGAN-GP-selected_G.pth'

my_model = FootprintsPredictor()
my_model.load_model(model_file)
my_model.eval()

predmap = my_model.test_single_im(frames[reference_frame_idx]['im'])
```

```python
import matplotlib.pyplot as plt
predmap[predmap<0.] = 0
plt.figure(figsize=(25,20));plt.imshow(predmap);plt.axis('off');plt.show()
```
